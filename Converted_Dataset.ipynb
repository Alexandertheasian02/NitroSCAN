```python
from google.colab import drive
from PIL import Image
import os

# Mount Google Drive
drive.mount('/content/drive')

# Define the directory path
directory = '/content/drive/MyDrive/LCC/Dataset/Level 02'
destination_directory = '/content/drive/MyDrive/LCC/Resized_Images/Level 02'

# Create the destination directory if it doesn't exist
if not os.path.exists(destination_directory):
    os.makedirs(destination_directory)

# List files in the directory
files = os.listdir(directory)

# List files already resized in the destination directory
resized_files = os.listdir(destination_directory)
resized_files = [os.path.splitext(file)[0] for file in resized_files]

# Print each file in the directory
for file in files:
    file_path = os.path.join(directory, file)
    original = Image.open(file_path)
    resized = original.resize((256,256))
    #print(file)

    # Get the filename without extension
    filename_without_extension = os.path.splitext(file)[0]

    # Check if the resized image already exists in the destination directory
    if filename_without_extension in resized_files:
        print(f"Resized image already exists: {filename_without_extension}")
        continue

    # Define the destination file path with the new format
    destination_file_path = os.path.join(destination_directory, f"{filename_without_extension}.png")

    # Save the resized image to the destination directory in PNG format
    resized.save(destination_file_path, format="PNG")

    print(f"Resized image saved: {destination_file_path}")
```
